### ✅ **1. What is regularization in regression, and why is it important?**

**Regularization** is a technique used to **reduce model complexity** and **prevent overfitting** by adding a **penalty term** to the loss function of regression models.

#### 🔧 Why it matters:

* Prevents the model from learning noise in training data.
* Controls the size of coefficients to avoid excessive weights.
* Encourages **simpler, more generalizable models**.

In linear regression, the regularized loss function looks like:

* **Ridge (L2)**:

  $$
  \text{Loss} = \text{MSE} + \lambda \sum_{j=1}^p \beta_j^2
  $$
* **Lasso (L1)**:

  $$
  \text{Loss} = \text{MSE} + \lambda \sum_{j=1}^p |\beta_j|
  $$

---

### ✅ **2. Explain the difference between L1 and L2 regularization.**

| Feature                | **L1 (Lasso)**                                  | **L2 (Ridge)**                             |   |                  |
| ---------------------- | ----------------------------------------------- | ------------------------------------------ | - | ---------------- |
| Penalty term           | (\sum                                           | \beta\_j                                   | ) | $\sum \beta_j^2$ |
| Effect on coefficients | Can shrink some to **zero** (feature selection) | Shrinks all coefficients (no zeroing)      |   |                  |
| Sparse model           | Yes                                             | No                                         |   |                  |
| Convexity              | Convex (but not differentiable at 0)            | Convex and differentiable                  |   |                  |
| When to use            | Feature selection, sparse data                  | Multicollinearity, all features are useful |   |                  |

---

### ✅ **3. How does regularization help prevent overfitting in regression models?**

Overfitting happens when the model fits **noise and fluctuations** in the training data. Regularization:

* Penalizes **large coefficients**, making the model less sensitive to minor fluctuations.
* Helps in **generalization** by simplifying the hypothesis function.
* Forces the model to **focus on most relevant features** (especially in Lasso).

---

### ✅ **4. What is the regularization parameter in Ridge and Lasso regression, and how does it affect the model?**

The **regularization parameter** is typically denoted by $\lambda$ or `alpha` in scikit-learn.

* **Small $\lambda$**: Model behaves like normal regression (more flexible, more risk of overfitting).
* **Large $\lambda$**: Increases the penalty ⇒ shrinks coefficients ⇒ less overfitting but may underfit.

👉 **Hyperparameter tuning** is used to find the best value of $\lambda$.

---

### ✅ **5. What are Lasso, Ridge, and Elastic Net regression techniques?**

| Type            | Description                                                                                           |
| --------------- | ----------------------------------------------------------------------------------------------------- |
| **Lasso (L1)**  | Performs regularization **and** feature selection by shrinking some coefficients to **exactly zero**. |
| **Ridge (L2)**  | Penalizes large coefficients but doesn’t eliminate them; good when all features are useful.           |
| **Elastic Net** | Combines L1 and L2 penalties:                                                                         |

$$
\text{Loss} = \text{MSE} + \lambda_1 \sum |\beta_j| + \lambda_2 \sum \beta_j^2
$$

Ideal when you want **both feature selection and stability**.

---

### ✅ **6. Explain the differences between Lasso and Ridge regression.**

| Criteria          | Lasso                                    | Ridge                               |    |                     |
| ----------------- | ---------------------------------------- | ----------------------------------- | -- | ------------------- |
| Penalty Type      | L1 ((\sum                                | \beta                               | )) | L2 ($\sum \beta^2$) |
| Feature Selection | Yes (can set coefficients to 0)          | No (all features retained)          |    |                     |
| Use Case          | Sparse features, high dimensionality     | Collinearity, numerical stability   |    |                     |
| Performance       | Better when some features are irrelevant | Better when all features contribute |    |                     |

---

### ✅ **7. How does Elastic Net combine Lasso and Ridge?**

Elastic Net introduces a **mixing parameter $\alpha$** and balances:

* **L1 penalty (Lasso)**: feature selection
* **L2 penalty (Ridge)**: handles multicollinearity

Useful when:

* Features are **correlated**
* You want **both robustness** and **sparsity**

---

### ✅ **8. When would you prefer using Elastic Net over Lasso or Ridge?**

Use **Elastic Net** when:

* You have **many features**, some correlated
* Lasso alone is **too aggressive** (drops correlated predictors)
* Ridge doesn’t perform **feature selection**
* You need a **balance** between stability and sparsity

### ✅ **9. What is cross-validation, and why is it important in model evaluation?**

**Cross-validation** is a technique to assess how well a model generalizes to unseen data by splitting the data into **training and validation folds**.

#### 🧠 Why it matters:

* Reduces **variance** in performance estimation.
* Helps avoid **overfitting to the train-test split**.
* Common type: **K-Fold Cross-Validation** – data is divided into *K* subsets; the model is trained on *K-1* and validated on the remaining one, repeated *K* times.

### ✅ **16. What is Logistic Regression, and how is it different from Linear Regression?**

**Logistic Regression** is a **classification algorithm**, while **Linear Regression** is used for **regression** (predicting continuous values).

| Feature       | Logistic Regression                 | Linear Regression                     |
| ------------- | ----------------------------------- | ------------------------------------- |
| Output        | Probability → class label (0 or 1)  | Continuous numeric output             |
| Function used | **Sigmoid (logistic) function**     | Linear function                       |
| Use case      | Binary or multiclass classification | Predicting values like prices, scores |
| Loss function | Log-loss (Cross-Entropy)            | Mean Squared Error (MSE)              |

**Key difference**: Logistic Regression uses the **logistic function** to model **P(y=1 | x)** and outputs a probability between 0 and 1.

### ✅ **17. Explain the concept of the logistic (sigmoid) function and its role in Logistic Regression**

The **logistic (sigmoid) function** maps any real-valued number to a value between **0 and 1**, making it ideal for modeling probabilities.

#### 🧮 **Sigmoid function formula**:

$$
\sigma(z) = \frac{1}{1 + e^{-z}}, \quad \text{where } z = \beta_0 + \beta_1 x_1 + \dots + \beta_n x_n
$$

#### 🔍 Role in Logistic Regression:

* Takes the linear combination of features ($z$) and **converts it into a probability**.
* If $\sigma(z) \ge 0.5$, the model predicts **class 1**, else **class 0**.
* It allows us to interpret the output as:

  $$
  P(y=1 \mid x) = \sigma(z)
  $$

#### 📈 Sigmoid curve characteristics:

* S-shaped
* As $z \to \infty$, $\sigma(z) \to 1$; as $z \to -\infty$, $\sigma(z) \to 0$
* Threshold-based decision making (usually 0.5)

### ✅ **18. What are the assumptions of Logistic Regression?**

1. **Binary or categorical outcome**: Dependent variable should be binary (or multinomial in extended versions).
2. **Independence**: Observations must be independent of each other.
3. **Linearity of log-odds**: Logistic regression assumes a **linear relationship** between the logit of the outcome and the independent variables.
4. **No multicollinearity**: Features shouldn’t be highly correlated.
5. **Large sample size**: More data helps stabilize the estimates.

---

### ✅ **19. How would you interpret the coefficients in a Logistic Regression model?**

Each coefficient $\beta_j$ represents the **log-odds change** in the outcome for a 1-unit increase in feature $x_j$, holding others constant.

* $\beta_j > 0$: Increases the probability of class 1
* $\beta_j < 0$: Decreases the probability of class 1
* Use `exp(β_j)` to interpret in terms of **odds ratio**.

### ✅ **20. How does regularization improve the performance of a Logistic Regression model?**

* Prevents **overfitting**, especially with many features or noise.
* Shrinks less relevant feature weights.
* Improves **generalization** on unseen data.
* Two types:

  * **L1 (Lasso)**: Promotes sparsity (feature selection)
  * **L2 (Ridge)**: Smooths all coefficients

### ✅ **24. Explain precision, recall, and F1-score**

| Metric    | Formula                                                                                   | Meaning                                           |
| --------- | ----------------------------------------------------------------------------------------- | ------------------------------------------------- |
| Precision | $\frac{TP}{TP + FP}$                                                                      | Out of predicted positives, how many were correct |
| Recall    | $\frac{TP}{TP + FN}$                                                                      | Out of actual positives, how many were found      |
| F1-Score  | $2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}$ | Harmonic mean of precision and recall             |

---

### ✅ **25. How would you use ROC and AUC to evaluate a classification model?**

* **ROC (Receiver Operating Characteristic)** plots **TPR vs FPR** at various thresholds.
* **AUC (Area Under Curve)** measures the model's ability to **distinguish between classes**.

  * AUC = 1: perfect model
  * AUC = 0.5: random guessing

```python

```

---

### ✅ **26. What is the confusion matrix and how is it used?**

A **2x2 table** summarizing predictions vs actual outcomes:

|                 | Predicted Positive  | Predicted Negative  |
| --------------- | ------------------- | ------------------- |
| Actual Positive | True Positive (TP)  | False Negative (FN) |
| Actual Negative | False Positive (FP) | True Negative (TN)  |

Used to compute accuracy, precision, recall, F1-score, etc.
